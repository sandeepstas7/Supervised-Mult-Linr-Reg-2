```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('50_Startups.csv')
data.head()
```

```python
X = data.iloc[:,:-1].values
y = data.iloc[:,4].values
X,y
```

```python
# Encoding categorical data
# Encoding the Independent variable
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
X[:,3] = labelencoder_X.fit_transform(X[:,3])
onehotencoder = OneHotEncoder(categorical_features=[3]) 
X = onehotencoder.fit_transform(X).toarray()
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
# Fitting multiple linear regression to the training set

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
# Prediction of the test set result(Performance of our regression  model on test data set)

y_pred = regressor.predict(X_test)
y_pred
```

```python
# Do u think it's actually the optimal model that we can make with the dataset that we have here ?
# When we build this model we actually used all the independent variables, but what if among this independent variables there
# are some highly statistically significant i.e., have great impact on the dependent variable and some of them are not
# statistically significant at all
# that means if we remove the non statistically significant variables from the model we'll still get some amazing predictions.

# So, here our goal is to actually find a team an optimal team of independent variables so each independent variable in the
# team has great impact(highly statistically significant) and definetly has an effect on the dependent variable.
# and this effect can be positive i.e,for an increase of 1 unit of the independent variable the dependent variable increase
# or it can be negative i.e,for a decrease of 1 unit of the independent variable the dependent variable decrease.
```

```python
#the principle of backward elemination is that we'll remove each independent variable that is not statistically significant
# (i.e., p-value > signif level(0.05)) one by one.

# S1 : Select a significant level to stay in the model (eg SL = 0.05)
# S2 : Fit the model with all possible predictors
# S3 : Consider the predictor with the highest P-value.If P > SL, go to S4, otherwise go to FIN
# S4 : Remove the predictor
# S5 : Fit model with out variable

# So we're  building the optimal model using "backward elimination"
```

```python
import statsmodels.formula.api as sm

# we just need to add a column of ones in our matrix of features or matrix of independent variables 
# And to understand why we need to do this go through mutiple linear equation(y = b0+b1x1+....+bnxn)
# we notice in this equation that there is constant(b0) which is not associated with independent variable like 'x'(i.e, x0) 
# but we can associate 'x0' to this constant 'b0'(its actually 'x0 = 1')
# And actually this statsmodels library that we r using doesn't take into account this 'b0' constant, so we'll somehow need to
# add it in our matrix of independent variables.
# In most libraries that we use to build a linear regression model its definetely included the fact that there is a constant(b0)
# in the multiple linear regression equation, but it's not actually the case for this stats model library which we'll use to
# compute p-values and evaluate statistically significance of our independent variable

# that's why we need do do something here and if something is to actually add this column of ones here in our matrix of features
# X = np.append(arr = X,values = np.ones((50,1)),axis = 1)

# to prevent a datatype error from happening we need to convert this array of 1's into integer(i.e,np.ones((50,1)).astype(int))
# X = np.append(arr = X,values = np.ones((50,1)).astype(int),axis = 1)

# if we run the above code it'll add the column of 1's at the end of matrix of features X,and we would like to have it at the 
# beginning so in place of "arr = X take  np.ones((50,1)).astype(int)" and in palce of "values take X"

X = np.append(arr = np.ones((50,1)).astype(int),values = X,axis = 1)
X
```

```python
X_opt = X[:,[0,1,2,3,4,5,6]]
regressor_OLS = sm.OLS(endog = y,exog = X_opt).fit()
regressor_OLS.summary()
```

```python
#while removing the variable with highest p-values go by index of array X 
#(constant(b0) with dummy varble(x0) with 1's has index 0 )
X_opt = X[:,[0,1,2,3,4,6]]
regressor_OLS = sm.OLS(endog = y,exog = X_opt).fit()
regressor_OLS.summary()
```

```python
X_opt = X[:,[0,1,2,3,4]]
regressor_OLS = sm.OLS(endog = y,exog = X_opt).fit()
regressor_OLS.summary()
```
